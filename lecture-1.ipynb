### Matplotlib is one of the most popular Python libraries used for data visualization. It helps you create graphs, charts, and plots so that data can be understood more easily.

**Matplotlib** allows us to visualize data through:

1. Line Charts – show trends over time
2. Bar Charts – compare different categories
3. Pie Charts – show percentages or proportions
4. Scatter Plots – show relationships between variables
5. Histograms – show data distribution

___
## Data Visualization :
When data is displayed in tables, it can be difficult to understand quickly.It is the process of representing data in a visual form such as graphs, charts, maps, or diagrams so that people can understand patterns, trends, and insights more easily.

___
## Pyplot :
Matplotlib is the full library in Python.It contains many modules and tools for creating all kinds of plots.Pyplot is a sub-module of Matplotlib which contains functions which provides a simple interface for creating plots quickly.It is the most commonly used part of Matplotlib.

___
## Some Important Terms in Matplotlib : 

| Term           | Short Definition                                                         |
| -------------- | ------------------------------------------------------------------------ |
| **Data Point** | A single value or observation plotted on a graph (x_value, y_value).                        |
| **X-Axis**     | The horizontal axis of a graph.                                          |
| **Y-Axis**     | The vertical axis of a graph.                                            |
| **Figure**     | The entire window or canvas containing one or more plots.                |
| **Axes**       | The plotting area inside a figure where data is displayed.               |
| **Plot**       | A graphical representation of data e.g.charts(line, pie , bar).                                      |
| **Marker**     | A symbol used to highlight individual data points (e.g., `o`, `*`, `x`). |
| **Line Style** | The appearance of a line (solid, dashed, dotted, etc.).                  |
| **Color**      | Specifies the color of plot elements such as lines and markers.          |
| **Legend**     | A box that identifies different data series in a plot.                   |
| **Label**      | Text describing an axis or data series.                                  |
| **Title**      | The heading that describes the graph.                                    |
| **Grid**       | Reference lines that make graph values easier to read.                   |

### Importing Matplotlib with some Object Name

```python
import matplotlib.pyplot as plt
```

___
### Creating a Plot :

```python
x = [1, 2 , 3, 4, 5]
y = [10, 20 ,15 ,25,18]

plt.plot(x,y)
plt.show()
```

___
### Creating Graph with Title and Labels :

```python
x = ["Mon" , "Tues", "Wed", "Thur", "Fri", "Sat", "Sun"]
y = [30, 50, 40, 70, 80, 10 , 100]

plt.plot(x,y)
plt.title("Sales Per Day")
plt.xlabel("Days")
plt.ylabel("Sales Percentage")

plt.show()
```

### Markers:

| Marker | Meaning        |
| ------ | -------------- |
| **o**  | Circle         |
| **.**  | Point          |
| **x**  | X marker       |
| **+**  | Plus sign      |
|    *  | Star           |
| **s**  | Square         |
| **D**  | Diamond        |

```python
plt.plot([1,2,3], [4,5,6],color='r',linestyle='--',marker='o')
plt.show()
```

### Some Additional Featues :

| Property            | Short Definition                                     | Example               |
| ------------------- | ---------------------------------------------------- | --------------------- |
| **linewidth (lw)**  | Controls thickness of the line                       | thick line            |
| **markersize (ms)** | Controls size of markers                             | large dots            |
| **label**           | Name used in legend                                  | “Sales Data”          |
| **alpha**           | Controls transparency of plot                        | semi-transparent line |
| **markerfacecolor** | Fill color of marker                                 | red-filled dots       |
| **markeredgecolor** | Border color of marker                               | black-bordered dots   |
| **fontsize**        | Changes size of text (title/labels)                  | big text              |
| **xlim**            | Sets range of X-axis                                 | limited x range       |
| **ylim**            | Sets range of Y-axis                                 | limited y range       |
| **grid**            | Shows background reference lines                     | grid lines visible    |


```python
x = [1, 2, 3, 4]
y = [10, 20, 15, 25]

plt.plot(x, y,
         color='blue',
         linestyle='--',
         marker='o',
         linewidth=2,
         markersize=8,
         alpha=0.7,
         label='Sales Data',
         markerfacecolor='yellow',
         markeredgecolor='black')

plt.title("Complete Matplotlib Example", fontsize=14)
plt.xlabel("X Axis", fontsize=12)
plt.ylabel("Y Axis", fontsize=12)

plt.xlim(0, 6)
plt.ylim(0, 30)

plt.grid(True)
plt.legend() #You can also pass upperleft or boottom right in bracket as parameter.

plt.show()
```

___
### plt.xticks() and plt.yticks() :
They are used to control the tick marks and labels on the X-axis and Y-axis.

```python
x = [1, 2, 3, 4]
y = [10, 20, 30, 40]

plt.plot(x, y, marker='o')

plt.xticks([1, 2, 3, 4], ["One", "Two", "Three", "Four"])
plt.yticks([10, 20, 30, 40], ["10%", "20%", "30%", "40%"])

plt.show()
```
